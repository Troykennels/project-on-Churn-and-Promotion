 PROMOTION PROJECT INSIGHTS
(Data Understanding â€¢ Employee Performance â€¢ Promotion Trends â€¢ Model Findings)
 1. Promotion Rate Is Very Low
Promotion distribution chart shows:
	â€¢	0 = Not Promoted â†’ VERY high count
	â€¢	1 = Promoted â†’ Very low count
This indicates a highly imbalanced dataset, meaning:
	â€¢	Only a small percentage of employees are promoted.
	â€¢	Promotions are selective and competitive.
	â€¢	HR likely uses strict criteria.
Implication:
Models must handle imbalance (e.g., class weights or SMOTE), from an HR standpoint, this suggests a narrow promotion funnel.
2. Division Strongly Influences Promotion
Division vs Promotion chart shows:
â€¢ Some divisions (e.g., Commercial, Customer Service) have many employees but few promotions.
â€¢	Smaller specialized divisions (like Research, IT Solutions, Strategy) show higher promotion ratios compared to their size.
Insight:
Promotion chances differ across departments, employees in strategic, technical, or specialized units appear more likely to be promoted.
Business Impact: Possible skill gaps or unequal opportunity distribution.

3. Qualification Plays a Major Role
From the dataset:
	â€¢	Higher degrees (MSc, MBA, PhD) correlate with higher promotion likelihood.
	â€¢	Lower qualifications (HND/First Degree) show fewer promotions.
	â€¢	Missing qualifications were rare but filled
 Insight: Organization seems to value advanced education in promotion decisions.
 Recommendation: Upskilling programs may improve promotion mobility.
 4. Performance Scores Are Critical
There are two important performance metrics:
1. Last Performance Score
2. Training Score Average
Employees with high training scores and high-performance review scores are significantly more likely to be promoted.
Insight: Promotion is performance-driven, not tenure-based.
 Recommendation: HR should monitor performance development programsâ€”these directly affect promotion readiness.
 5. Trainings Attended + Targets Met Matter
Employees who attend more trainings and meet targets consistently tend to have better promotion odds.
 Insight: The organization promotes employees who demonstrate continuous improvement and goal achievement.



 6. Year of Recruitment (Tenure Effect)
Employees hired in earlier years (older tenure) show slightly higher promotion frequency.
However, this is not linear â€” meaning promotion is not simply based on time spent, but is influenced along with performance and qualification.
 Insight: Experience is helpful but not enough alone.
 7. Foreign Schooled Employees Show Higher Promotion Odds
Based on the variables:
Foreign_schooled = Yes â†’ Higher likelihood as this may indicate perceived superior training or wider experience.
Insight:
The organization values global exposure / external education.

 8. Disciplinary Actions Reduce Promotion Chances
Employees with â€œPast_Disciplinary_Action = Yesâ€ have almost zero promotions.
 Insight: Behavioral history is a strong negative factor.

 9. Marital Status Has Minimal Impact
Marital status (Married/Single) does not significantly affect promotion.
 Insight: This is expected  HR typically avoids bias regarding marital status.
OVERALL SUMMARY INSIGHT
â€œPromotion in this company is largely determined by performance indicators, qualifications, training participation, and division type, with minimal influence from demographic variables.â€
Employees who score highly in:
1.	Performance reviews
2.	Training outcomes
3.	Target achievements
are most likely to be promoted.
Employees with:
1.	 lower education

1.	 HR should strengthen performance-improvement programs

```python
#encoding my categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in promotion_data.columns:
    if promotion_data[col].dtype == 'object':
        promotion_data[col] = le.fit_transform(promotion_data[col])
```

```python
# this is to check correlation between variables
promotion_data.corr()
```

```python
#slicing the data into features and target variable
x = promotion_data.drop('Promoted_or_Not', axis=1)
y = promotion_data['Promoted_or_Not']
```

```python
#splitting the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# i want to use RandomForestClassifier model and also fit the model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    max_depth=None
)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix heatmap
plt.figure(figsize=(10,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='Blues', fmt='d')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
import pandas as pd

importances = pd.Series(model.feature_importances_, index=x.columns)
importances.sort_values(ascending=True).plot(kind='barh', figsize=(10,6))

plt.title("Feature Importance")
plt.xlabel("Importance Score")
plt.show()

importances.sort_values(ascending=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
import joblib

# Load cleaned dataset
df = pd.read_csv("promotion_dataset.csv")

# Encode categorical columns
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = le.fit_transform(df[col])

# Features and target
X = df.drop("Promoted_or_Not", axis=1)
y = df["Promoted_or_Not"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Train Random Forest model
model = RandomForestClassifier(
    n_estimators=300, random_state=42
)
model.fit(X_train, y_train)

# Save model
joblib.dump(model, "promotion_model.pkl")
joblib.dump(le, "label_encoder.pkl")  # Save encoder (if needed)
```
